```python
import pandas as pd
```

```python
lafc_goals = pd.read_csv('lafc 2019 season - Goals Tracker.csv')
```

```python
lafc = pd.DataFrame(lafc_goals)
```

```python
lafc.head()
```

```python
import matplotlib.pyplot as plt
```

```python
lafc.set_index('goal')
```

```python
first_six = lafc[0:18]
```

```python
first_six.head()
```

```python
import seaborn as sns
```

```python
first_six.set_index('goal')
```

```python
goal_scorers = sns.countplot(x='goal_scorer', data=first_six, palette="Set3", order = first_six['goal_scorer'].value_counts().index)
plt.xticks(rotation=15)
```

```python
fig = goal_scorers.get_figure()
fig.savefig("lafc_first_six.png")
```
