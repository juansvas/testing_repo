```
# Set up your tools
from sklearn.datasets import load_boston
from pandas_profiling import ProfileReport
import pandas_profiling
import pandas as pd
```

```
# Load and check your data
boston = load_boston()
data = pd.DataFrame(boston.data, columns=boston.feature_names)
data.head()
```

```
# Run Pandas Profiling Report
profile = ProfileReport(data)
profile.to_file('boston_profile_report.html')
```

```
# Display report
import IPython
IPython.display.HTML(filename='/content/boston_profile_report.html')
```
